## Lab 3 - Bài 1

```python
# Code lab 3 bài 1 ở đây
day_so = []
while True:

    so = input("nhập số (nhập q để thoát): ")
    if so.lower() == "q":
        print("Thoát chương trình.")
        break   
    else:
        try:
            phan_tu = int(so)
            day_so.append(phan_tu)
            print("Bạn vừa nhập số: ",phan_tu)
        except ValueError:
            print("Không hợp lệ! Yêu cầu nhập q để thoát chương trình")

day_so.sort(reverse=True)
print("dãy số đã sắp xếp tăng dần")
print(day_so)
print("giá trị lớn nhất dãy số: ",day_so[0])
print("giá trị nhỏ nhất dãy số: ",day_so[-1])

Tong = 0
dem = 0
for phan_tu in day_so:
    if phan_tu % 3 == 0:
        Tong += phan_tu
        dem +=1

if dem != 0:
    print("Kết quả: ",Tong/dem)
else:
    print("Không có số nào chia hết cho 3 ")
```

## Lab 3 - Bài 5: bài tập làm thêm

```python
# code lab 3  bài làm thêm ở đây 
import math
import datetime

# Lưu lịch sử phép tính
lich_su = []

menu = """
-----MENU-----
0. Thoát chương trình
1. Phép tính cơ bản: Cộng, trừ, nhân, chia
2. Lũy thừa: Tính x^y
3. Căn bậc hai: Tính √x
4. Hàm lượng giác: Sin, Cos, Tan
5. Logarit: log10, ln, log cơ số tùy chọn
6. Giải phương trình bậc 1 (ax + b = 0)
7. Giải phương trình bậc 2 (ax^2 + bx + c = 0)
8. Lịch sử: Xem lại các phép tính
9. Thời gian: Hiển thị thời gian hiện tại
10. Giao diện thân thiện: Menu tiếng Việt
"""

while True:
    print(menu)
    lua_chon = input("Chọn chức năng chương trình: ")
    match lua_chon:
        case "0":
            print("Thoát chương trình.")
            break

        case "1":
            print("1. Phép tính cơ bản: Cộng, trừ, nhân, chia")
            a = float(input("Nhập a: "))
            b = float(input("Nhập b: "))
            kq = {
                "Cộng": a + b,
                "Trừ": a - b,
                "Nhân": a * b,
                "Chia": a / b if b != 0 else "Không chia cho 0"
            }
            for k, v in kq.items():
                print(f"{k}: {v}")
            lich_su.append(f"Phép tính cơ bản với {a}, {b} => {kq}")

        case "2":
            print("2. Lũy thừa: Tính x^y")
            x = float(input("Nhập x: "))
            y = float(input("Nhập y: "))
            kq = x**y
            print("Kết quả:", kq)
            lich_su.append(f"{x}^{y} = {kq}")

        case "3":
            print("3. Căn bậc hai: √x")
            x = float(input("Nhập x: "))
            if x >= 0:
                kq = math.sqrt(x)
                print("√x =", kq)
                lich_su.append(f"√{x} = {kq}")
            else:
                print("Không tính được căn bậc hai số âm!")

        case "4":
            print("4. Hàm lượng giác: sin, cos, tan")
            goc = float(input("Nhập góc (độ): "))
            rad = math.radians(goc)
            kq = {
                "sin": math.sin(rad),
                "cos": math.cos(rad),
                "tan": math.tan(rad) if math.cos(rad) != 0 else "Không xác định"
            }
            for k, v in kq.items():
                print(f"{k}({goc}°) = {v}")
            lich_su.append(f"Hàm lượng giác góc {goc}° => {kq}")

        case "5":
            print("5. Logarit")
            x = float(input("Nhập số x (>0): "))
            if x > 0:
                print("log10(x) =", math.log10(x))
                print("ln(x) =", math.log(x))
                co_so = float(input("Nhập cơ số: "))
                if co_so > 0 and co_so != 1:
                    print(f"log cơ số {co_so} của {x} =", math.log(x, co_so))
                else:
                    print("Cơ số không hợp lệ (phải >0, ≠1).")
            else:
                print("x phải > 0")

        case "6":
            print("6. Giải PT bậc 1: ax + b = 0")
            a = float(input("Nhập a: "))
            b = float(input("Nhập b: "))
            if a != 0:
                x = -b / a
                print("Nghiệm:", x)
                lich_su.append(f"PT bậc 1: {a}x+{b}=0 => x={x}")
            else:
                print("Phương trình vô nghiệm (a=0)!")

        case "7":
            print("7. Giải PT bậc 2: ax^2 + bx + c = 0")
            a = float(input("Nhập a: "))
            b = float(input("Nhập b: "))
            c = float(input("Nhập c: "))
            if a == 0:
                print("Đây không phải PT bậc 2.")
            else:
                delta = b**2 - 4*a*c
                if delta > 0:
                    x1 = (-b + math.sqrt(delta)) / (2*a)
                    x2 = (-b - math.sqrt(delta)) / (2*a)
                    print("2 nghiệm phân biệt:", x1, x2)
                elif delta == 0:
                    x = -b / (2*a)
                    print("Nghiệm kép:", x)
                else:
                    print("PT vô nghiệm thực")

        case "8":
            print("=== Lịch sử các phép tính ===")
            if lich_su:
                for item in lich_su:
                    print("-", item)
            else:
                print("Chưa có phép tính nào!")

        case "9":
            print("9. Thời gian hiện tại:")
            now = datetime.datetime.now()
            print(now.strftime("%Y-%m-%d %H:%M:%S"))

        case "10":
            print("10. Menu tiếng Việt đã hiển thị sẵn ở trên.")

        case _:
            print("Vui lòng chọn đúng chức năng (0-10).")

```
